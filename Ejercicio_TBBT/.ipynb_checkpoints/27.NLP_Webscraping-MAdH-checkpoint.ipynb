## Ejercicio NLP & WebScraping

1. Hacer Web Scraping sobre los siguientes links para obtener los diálogos de cada capítulo de la serie:

- https://bigbangtrans.wordpress.com/series-1-episode-1-pilot-episode/
- https://bigbangtrans.wordpress.com/series-1-episode-2-the-big-bran-hypothesis/
- https://bigbangtrans.wordpress.com/series-1-episode-3-the-fuzzy-boots-corollary/
- https://bigbangtrans.wordpress.com/series-1-episode-4-the-luminous-fish-effect/
- https://bigbangtrans.wordpress.com/series-1-episode-5-the-hamburger-postulate/

2. Una vez que se obtuvieron los diálogos de cada capítulo, unificamos los textos de cada capítulo en uno solo.
3. El primer paso es separar los diálogos por personaje (Sheldon, Leonard, Penny, Raj, Howard). 
4. Tokenizar cada dialogo por personaje con la función **`.word_tokenize()`**. 
5. Eliminar stopwords y signos de puntuación (. , : ; ¡ ¿ ? …) de las listas con tokens
6. Encontrar la riqueza léxica de cada personaje
7. Encontrar las 3 palabras más utilizadas por cada personaje
8. Encontrar palabras similares para la palabra más utilizada de cada personaje
9. Encontrar hápax de cada personaje
10. Utilizar la función nltk.FreqDist() para mostrar en una gráfica las 20 palabras que más se utilizan.
11. Repetir el ejercicio con todos los episodios de la serie.

```python
import requests
from bs4 import BeautifulSoup

import pandas as pd

import re

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.text import Text
from nltk import FreqDist

import string

from collections import Counter

import matplotlib.pyplot as plt
```

### 1. Extraer los textos

```python
#URLs de las 5 páginas para scrappear
urls = [
    'https://bigbangtrans.wordpress.com/series-1-episode-1-pilot-episode/',
    'https://bigbangtrans.wordpress.com/series-1-episode-2-the-big-bran-hypothesis/',
    'https://bigbangtrans.wordpress.com/series-1-episode-3-the-fuzzy-boots-corollary/',
    'https://bigbangtrans.wordpress.com/series-1-episode-4-the-luminous-fish-effect/',
    'https://bigbangtrans.wordpress.com/series-1-episode-5-the-hamburger-postulate/'
]

#Definimos una función para extraer y guardar el contenido de un capítulo
def scrapping(url, capitulo):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    entrytext = soup.find("div", class_="entrytext")
    
    if entrytext:
        parrafos = entrytext.find_all("p")
        dialogo = "\n".join([p.text for p in parrafos])
        
        archivo = f"Ejercicio_TBBT/Capitulo_{capitulo}.txt"
        
        with open(archivo, "w", encoding="utf-8") as file: #Guardamos el diálogo en un archivo de texto
            file.write(dialogo)
        print(f"Capítulo {capitulo} guardado en {archivo}")
    else:
        print(f"No se pudo encontrar el contenido en {url}")

#Iteramos a través de las URLs para extraer los diálogos
for i, url in enumerate(urls, start=1):
    scrapping(url, i)
```

### 2. Unificar los textos

```python
#Creamos una lista para almacenar el contenido de cada archivo
textos = []

#Leemos cada archivo y agregamos su contenido a la lista
for i in range(1, 6):
    archivo = f"Ejercicio_TBBT/Capitulo_{i}.txt"
    try:
        with open(archivo, "r", encoding="utf-8") as file:
            texto_capitulo = file.read()
            textos.append(texto_capitulo)
    except FileNotFoundError:
        print(f"El archivo {archivo} no se encontró.")

#Unificamos todos los textos en uno solo
texto_unico = "\n\n".join(textos)

#Guardamos el texto unificado en un archivo
archivo_unico = f"Ejercicio_TBBT/Capitulos_1-5.txt"
with open(archivo_unico, "w", encoding="utf-8") as file:
    file.write(texto_unico)

print(f"Los textos se han unido en {archivo_unico}")
```

### 3. Separar los diálogos

```python
# Ruta del archivo unificado
archivo_unico = "Capitulos_1-5.txt"

# Leer el archivo unificado
with open(archivo_unico, "r", encoding="utf-8") as file:
    texto_unificado = file.read()

# Eliminar texto entre paréntesis
texto_sin_parentesis = re.sub(r'\([^)]*\)', '', texto_unificado)
```

```python
# Dividir los diálogos por línea
lineas = texto_sin_parentesis.split('\n')

# Inicializar un diccionario para almacenar los diálogos por personaje
dialogos_por_personaje = {}

# Recorrer las líneas y separar los diálogos por personaje
personaje_actual = None
for linea in lineas:
    if ":" in linea:
        partes = linea.split(":")
        personaje_actual = partes[0].strip()
        dialogo = ":".join(partes[1:]).strip()
        if personaje_actual in dialogos_por_personaje:
            dialogos_por_personaje[personaje_actual].append(dialogo)
        else:
            dialogos_por_personaje[personaje_actual] = [dialogo]
    elif personaje_actual is not None:
        # Si la línea no contiene un ":" y tenemos un personaje actual,
        # la agregamos como continuación de su diálogo
        dialogos_por_personaje[personaje_actual][-1] += f"\n{linea.strip()}"
```

```python
# Lista de personajes permitidos
personajes_permitidos = ["Sheldon", "Leonard", "Penny", "Raj", "Howard"]

# Crear un nuevo diccionario solo con los personajes permitidos
dialogos_filtrados = {key: value for key, value in dialogos_por_personaje.items() if key in personajes_permitidos}
```

### 4. Tokenizar los diálogos

```python
# Reemplazar "’" por "'"
for personaje, dialogos in dialogos_filtrados.items():
    dialogos_filtrados[personaje] = [dialogo.replace("’", "'") for dialogo in dialogos]

# Inicializar el diccionario para los diálogos tokenizados
dialogos_tokenizados = {}

# Tokenizar los diálogos de cada personaje
for personaje, dialogos in dialogos_filtrados.items():
    dialogos_tokenizados[personaje] = [word_tokenize(dialogo, language='english') for dialogo in dialogos]
```

### 5. Eliminar stepwords y signos

```python
# Lista de stopwords en inglés
stopwords = nltk.corpus.stopwords.words("english")

# Lista de signos de puntuación que deseamos eliminar
punctuation = [".", ",", ":", ";", "!", "?", "…"]

# Función para limpiar los tokens de un diálogo
def limpiar_tokens(tokens):
    tokens_limpios = [token for token in tokens if token.lower() not in stopwords and token not in punctuation]
    return tokens_limpios
```

```python
# Limpiar los tokens de todos los diálogos en el diccionario
dialogos_tokenizados_limpios = {}
for personaje, tokens_dialogo in dialogos_tokenizados.items():
    dialogos_limpios = [limpiar_tokens(tokens) for tokens in tokens_dialogo]
    dialogos_tokenizados_limpios[personaje] = dialogos_limpios

# Función para eliminar palabras con apóstrofes
def eliminar_palabras_con_apostrofe(dialogos_tokenizados):
    dialogos_sap = {}
    for personaje, dialogos in dialogos_tokenizados.items():
        dialogos_sap[personaje] = [
            [palabra for palabra in dialogo if "'" not in palabra]  # Filtrar palabras sin apóstrofes
            for dialogo in dialogos
        ]
    return dialogos_sap

# Aplicar la función para eliminar palabras con apóstrofes
dialogos_limpios = eliminar_palabras_con_apostrofe(dialogos_tokenizados_limpios)
```

### 6. Encontrar riqueza léxica

```python
# Crear un diccionario para almacenar la riqueza léxica de cada personaje
riqueza_lexica = {}

# Calcular la riqueza léxica para cada personaje
for personaje, dialogo in dialogos_limpios.items():
    # Obtener una lista plana de todas las palabras en los diálogos del personaje
    palabras_totales = [palabra for frase in dialogo for palabra in frase]
    # Calcular el número de palabras únicas
    palabras_unicas = set(palabras_totales)
    # Calcular la riqueza léxica (número de palabras únicas / número total de palabras)
    riqueza = len(palabras_unicas) / len(palabras_totales)
    riqueza_lexica[personaje] = riqueza

# Mostrar la riqueza léxica de cada personaje
for personaje, riqueza in riqueza_lexica.items():
    print(f"Riqueza léxica de {personaje}: {riqueza:.2f}")
```

### 7. Encontrar 3 palabras más utilizadas

```python
# Crear un diccionario para almacenar las 3 palabras más utilizadas de cada personaje
palabras_mas_utilizadas = {}

# Encontrar las 3 palabras más utilizadas para cada personaje
for personaje, dialogo in dialogos_limpios.items():
    # Obtener una lista plana de todas las palabras en los diálogos del personaje
    palabras_totales = [palabra for frase in dialogo for palabra in frase]
    # Contar la frecuencia de cada palabra
    frecuencia_palabras = Counter(palabras_totales)
    # Obtener las 3 palabras más utilizadas
    palabras_top = frecuencia_palabras.most_common(3)
    palabras_mas_utilizadas[personaje] = palabras_top

# Mostrar las 3 palabras más utilizadas de cada personaje
for personaje, palabras_top in palabras_mas_utilizadas.items():
    print(f"Las 3 palabras más utilizadas de {personaje}:")
    for palabra, frecuencia in palabras_top:
        print(f"{palabra}: {frecuencia} veces")
    print()
```

### 8. Encontrar similares a la más utilizada

```python
# Transformar los objetos string en texto
dialogos_en_texto = {}
for personaje, dialogos in dialogos_limpios.items():
    textos = [Text(dialogo) for dialogo in dialogos]
    dialogos_en_texto[personaje] = textos
```

```python
# Inicializar el diccionario para almacenar la palabra más utilizada por cada personaje
palabra_mas_utilizada = {}

# Encontrar la palabra más utilizada para cada personaje
for personaje, dialogos in dialogos_limpios.items():
    # Unir todos los diálogos de un personaje en una lista única de palabras
    todas_las_palabras = [palabra for dialogo in dialogos for palabra in dialogo]
    
    # Calcular la frecuencia de cada palabra
    frecuencia_palabras = FreqDist(todas_las_palabras)
    
    # Encontrar la palabra más común
    palabra_mas_comun = frecuencia_palabras.most_common(1)[0][0]
    
    # Almacenar la palabra más común en el diccionario palabra_mas_utilizada
    palabra_mas_utilizada[personaje] = palabra_mas_comun

# Imprimir la palabra más utilizada por cada personaje
for personaje, palabra in palabra_mas_utilizada.items():
    print(f"La palabra más utilizada por {personaje} es: {palabra}")
```

```python
# Convertir el diccionario a una lista de tuplas
lista_palabras_mas_utilizadas = [(personaje, palabra) for personaje, palabra in palabra_mas_utilizada.items()]

# Imprimir la lista de tuplas
print("Lista de palabras más utilizadas por personaje:")
for tupla in lista_palabras_mas_utilizadas:
    print(f"{tupla[0]}: {tupla[1]}")
```

```python
# Inicializar un diccionario para almacenar palabras similares
palabras_similares = {}

# Encontrar palabras similares para cada personaje
for personaje, palabra_mas_utilizada in lista_palabras_mas_utilizadas:
    # Convertir los diálogos limpios del personaje en una lista de palabras
    palabras_personaje = [palabra for dialogo in dialogos_limpios[personaje] for palabra in dialogo]
    
    # Convertir la lista de palabras en un objeto Text de NLTK
    texto_personaje = nltk.text.ContextIndex(palabras_personaje)
    
    # Encontrar palabras similares utilizando la función similar_words()
    similares = texto_personaje.similar_words(palabra_mas_utilizada)
    
    # Almacenar las palabras similares en el diccionario palabras_similares
    palabras_similares[personaje] = similares

# Imprimir palabras similares para cada personaje
for personaje, similares in palabras_similares.items():
    print(similares)
```

### 9. Encontrar hápax de cada personaje

```python
# Inicializar un diccionario para almacenar los hapax legomena de cada personaje
hapax_por_personaje = {}

# Encontrar los hapax legomena para cada personaje
for personaje, dialogos in dialogos_limpios.items():
    # Unir todos los diálogos del personaje en una lista única de palabras
    todas_las_palabras = [palabra for dialogo in dialogos for palabra in dialogo]
    
    # Calcular la frecuencia de cada palabra
    frecuencia_palabras = FreqDist(todas_las_palabras)
    
    # Encontrar las palabras que aparecen solo una vez (hapax legomena)
    hapax = [palabra for palabra, frecuencia in frecuencia_palabras.items() if frecuencia == 1]
    
    # Almacenar los hapax legomena en el diccionario hapax_por_personaje
    hapax_por_personaje[personaje] = hapax

# Imprimir los hapax legomena para cada personaje
for personaje, hapax in hapax_por_personaje.items():
    print(f"Hapax para {personaje}:")
    print(hapax)
```

### 10. Gráfica de las 20 palabras más utilizadas

```python
import matplotlib.pyplot as plt

# Unir todas las palabras de todos los diálogos en una lista única
todas_las_palabras = [palabra for dialogos in dialogos_limpios.values() for dialogo in dialogos for palabra in dialogo]

# Calcular la frecuencia de cada palabra en el conjunto de diálogos
frecuencia_palabras = nltk.FreqDist(todas_las_palabras)

# Obtener las 20 palabras más comunes
palabras_mas_comunes = frecuencia_palabras.most_common(20)

# Separar las palabras y las frecuencias en listas separadas
palabras, frecuencias = zip(*palabras_mas_comunes)

# Crear la gráfica de barras
plt.figure(figsize=(12, 6))
plt.bar(palabras, frecuencias)
plt.title("Las 20 palabras más utilizadas en los diálogos")
plt.xlabel("Palabras")
plt.ylabel("Frecuencia")
plt.xticks(rotation=45, fontsize=10)
plt.tight_layout()
plt.show()
```

### 10. Repetir con todos los episodios

```python
import pandas as pd

# URL de la página
url = "https://bigbangtrans.wordpress.com/series-1-episode-1-pilot-episode/"

# Realizar la solicitud GET a la URL
response = requests.get(url)

# Verificar si la solicitud fue exitosa
if response.status_code == 200:
    # Parsear el contenido HTML de la página con BeautifulSoup
    soup = BeautifulSoup(response.text, "html.parser")
    
    # Encontrar el div con id "sidebar"
    sidebar = soup.find(id="sidebar")
    
    # Encontrar todas las etiquetas <li> dentro de "sidebar"
    etiquetas_li = sidebar.find_all("li")
    
    # Inicializar una lista para almacenar las URLs
    urls = []
    
    # Recorrer las etiquetas <li> y extraer las URLs
    for etiqueta_li in etiquetas_li:
        # Ignorar la etiqueta con el nombre "About"
        if "About" not in etiqueta_li.text:
            # Obtener la URL del atributo "href"
            url = etiqueta_li.find("a")["href"]
            urls.append(url)
    
    # Crear un DataFrame de pandas con las URLs
    df = pd.DataFrame(urls, columns=["URLs"])
    
    # Guardar el DataFrame en un archivo CSV
    df.to_csv("Ejercicio_TBBT/urls_TBBT.csv", index=False)
    
    print("URLs extraídas y guardadas en 'urls_TBBT.csv'.")
else:
    print("No se pudo acceder a la página web.")
```

```python
import os

# Leer el archivo CSV con las URLs
df = pd.read_csv("Ejercicio_TBBT/urls_TBBT.csv")

# Carpeta donde se guardarán los archivos de texto de los diálogos
carpeta_destino = "Ejercicio_TBBT"

# Crear la carpeta de destino si no existe
os.makedirs(carpeta_destino, exist_ok=True)

# Recorrer las URLs y extraer los diálogos de cada capítulo
for index, row in df.iterrows():
    url = row["URLs"]
    
    # Realizar la solicitud GET a la URL
    response = requests.get(url)
    
    # Verificar si la solicitud fue exitosa
    if response.status_code == 200:
        # Parsear el contenido HTML de la página con BeautifulSoup
        soup = BeautifulSoup(response.text, "html.parser")
        
        # Encontrar el div con la clase "entrytext"
        entrytext = soup.find(class_="entrytext")
        
        # Encontrar todos los párrafos <p> con <span> dentro de "entrytext"
        parrafos = entrytext.find_all("p")
        
        # Inicializar una lista para almacenar los diálogos de cada capítulo
        dialogos = []
        
        # Recorrer los párrafos y extraer el texto de los <span>
        for parrafo in parrafos:
            spans = parrafo.find_all("span")
            texto_parrafo = " ".join(span.get_text() for span in spans)
            dialogos.append(texto_parrafo)
        
        # Crear un archivo de texto para el capítulo actual
        nombre_archivo = f"Capitulo_{index + 1}.txt"
        ruta_archivo = os.path.join(carpeta_destino, nombre_archivo)
        
        # Escribir los diálogos en el archivo de texto
        with open(ruta_archivo, "w", encoding="utf-8") as archivo:
            for dialogo in dialogos:
                archivo.write(dialogo + "\n")
        
        print(f"Diálogos del capítulo {index + 1} extraídos y guardados en '{nombre_archivo}'")
    else:
        print(f"No se pudo acceder a la página web: {url}")
```
